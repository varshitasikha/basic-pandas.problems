### add a new column called title_length that contains the lengh of the each title

```python
df['title_length'] = df['title'].str.len()
print(df['title_length'])
```

### create a column called year_added extracted from the date_added column

```python
# Strip leading/trailing spaces
df['date_added'] = df['date_added'].str.strip()

# Convert to datetime, coercing errors to NaT (missing datetime)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Now extract the year
df['year_added'] = df['date_added'].dt.year
```
